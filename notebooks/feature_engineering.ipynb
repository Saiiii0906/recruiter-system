```python
import pandas as pd
from pathlib import Path
```

```python
DATA_PATH = (
    Path.cwd().parent
    / "data"
    / "raw"
    / "India_runs_data_and_ai_challenge"
)

sample_df = pd.read_json(
    DATA_PATH / "sample_candidates.json"
)

print(sample_df.shape)
```

```python
candidate = sample_df.iloc[0]

candidate["candidate_id"]
```

```python
profile = candidate["profile"]

semantic_profile = f"""
Headline:
{profile.get("headline","")}

Summary:
{profile.get("summary","")}

Current Title:
{profile.get("current_title","")}

Industry:
{profile.get("current_industry","")}
"""

print(semantic_profile)
```

```python
skills = candidate["skills"]

skill_names = []

for skill in skills:
    skill_names.append(skill["name"])

skill_names[:20]
```

```python
signals = candidate["redrob_signals"]

behavior_score = (
    signals["profile_completeness_score"] * 0.15
    + signals["github_activity_score"] * 0.20
    + signals["interview_completion_rate"] * 100 * 0.25
    + signals["offer_acceptance_rate"] * 100 * 0.20
    + signals["recruiter_response_rate"] * 100 * 0.20
)

print(round(behavior_score,2))
```

```python
candidate_features = {
    "candidate_id": candidate["candidate_id"],
    "experience_years": profile["years_of_experience"],
    "skill_count": len(skills),
    "behavior_score": round(behavior_score,2),
    "semantic_profile": semantic_profile
}

candidate_features
```
